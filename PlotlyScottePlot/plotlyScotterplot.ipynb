```python
import numpy as np
import plotly.express as px
import logging

''' 
    @Author: v sanjay kumar
    @Date: 2024-08-13 04:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-13 04:00:30.
    @Title :program to draw a scatter plot for random 1000 x and y coordinates
'''

logging.basicConfig(filename='app5.log', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def plot_scatter(x, y):
    '''
    Function takes the x, y numpy arrays and plots the scatter plot

    Parameters:
    x (numpy array): Array of x coordinates
    y (numpy array): Array of y coordinates
    
    Returns:
    None
    '''
    
    fig = px.scatter(x=x, y=y, title='Scatter Plot of Random 1000 x and y Coordinates')
    logging.info("Plotting ")
    fig.update_layout(xaxis_title='x', yaxis_title='y')
    fig.show()

def main():
    try:
        x = np.random.randint(0, 1000, 1000)
        y = np.random.randint(0, 1000, 1000)
        plot_scatter(x, y)
    except Exception as e:
        logging.error(f"Error generating random data: {e}")

if __name__ == "__main__":
    main()
```

4. Write a program to draw a scatter plot for a given dataset and show datalabels on hover
https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv

```python
import numpy as np
import plotly.express as px
import logging
import pandas as pd

''' 
    @Author: v sanjay kumar
    @Date: 2024-08-13 04:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-13 04:00:30.
    @Title :program to draw line and scatter plots for random 500 x and y coordinates
'''

logging.basicConfig(filename='app6.log', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def plot_scatter(url):
    '''
    Function takes the URL of a CSV file and plots the scatter plot

    Parameters:
    url (str): The URL of the dataset
    
    Returns:
    None
    '''
    try:
        df = pd.read_csv(url)
        logging.info('Dataset loaded successfully.')

        fig = px.scatter(data_frame=df, x='Rank', y='Population', title='Scatter Plot of Population by Rank')
        logging.info('Scatter plot created successfully.')

        fig.update_layout(xaxis_title='Rank', yaxis_title='Population')
        fig.show()
        logging.info('Plot displayed successfully.')

    except Exception as e:
        logging.error(f"Error occurred: {e}")

def main():
    try:
        url = 'https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv'
        plot_scatter(url)
    except Exception as e:
        logging.error(f"Error in main: {e}")

if __name__ == "__main__":
    main()
```
