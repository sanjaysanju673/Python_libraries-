```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import logging
```

1. Write a program to draw box plot of life expectancy by continent for a data set given in a
url

```python
logging.basicConfig(filename='app1.log', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def create_box_plot(url):
    '''
    Function takes the parameter from the url and plots the box plot

    Parameters:
    url (str): The URL of the dataset

    Returns:
    None
    '''
    try:
        gapminder = pd.read_csv(url)
        logging.info('Dataset loaded successfully.')

        sns.boxplot(x='continent', y='lifeExp', data=gapminder)
        logging.info('Box plot created successfully.')

        plt.xlabel('Continent')
        plt.ylabel('Life Expectancy')
        plt.title('Box Plot of Life Expectancy by Continent')
        logging.info('Labels and title added to the plot.')

        plt.show()
        logging.info('Plot displayed successfully.')

    except Exception as e:
        logging.error(f'Error occurred: {e}')

def main():
    try:
        url = 'https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv'
        create_box_plot(url)
    
    except Exception as e:
        logging.error(f'Error occurred in main: {e}')

if __name__ == '__main__':
    main()
```

2. Write a program to draw a box plot of day by tips for a dataset given in a url

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import logging

''' 
    @Author: v sanjay kumar
    @Date: 2024-08-13 06:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-13 06:00:30.
    @Title : Program to draw a box plot of day by tips for a dataset given in a url
'''
logging.basicConfig(filename='raja.log', level=logging.INFO,format='%(asctime)s - %(levelname)s - %(message)s')
def plot_box(url):
    ''' Fuction takes the url from the main function and plot the box plot
     
    parameters-url

    return-None   
    '''
    df=pd.read_csv(url)
    logging.info('Data set created successfully')
    sns.boxplot(x='day',y='tip',color='green',data=df)
    logging.info("the box plot was created successfully")
    plt.xlabel("Days")
    plt.ylabel("Tips")
    plt.title("box plot of day by tips")



def main():
    logging.info('The function stats with here')
    try:
        url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
        plot_box(url)
    except Exception as e:
        logging.error('thise is error message',e)
        
if __name__ == '__main__':
    main()
```
