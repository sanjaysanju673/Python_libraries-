1. Write a program to draw a swarm plot of total bill against size for a given dataset


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import logging
''' 
    @Author: v sanjay kumar
    @Date: 2024-08-13 04:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-13 04:00:30.
    @Title :program to draw a swarm plot of total bill against size for a given dataset.
'''


logging.basicConfig(filename='app1.log', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def create_swarm_plot(url):
    '''
    Function takes the parameter from the url and plots the swarm plot

    Parameters:
    url (str): The URL of the dataset

    Returns:
    None
    '''
    try:
        df = pd.read_csv(url)
        logging.info('Dataset loaded successfully.')

        plt.figure(figsize=(10, 6))
        sns.swarmplot(x='size', y='total_bill', data=df)
        logging.info('Swarm plot created successfully.')

        
        plt.xlabel('Size')
        plt.ylabel('Total Bill')
        plt.title('Swarm Plot of Total Bill by Size')
        logging.info('Labels and title added to the plot.')

       
        plt.show()
        logging.info('Plot displayed successfully.')

    except Exception as e:
        logging.error(f'Error occurred: {e}')

def main():
    try:
        url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
        create_swarm_plot(url)
    
    except Exception as e:
        logging.error(f'Error occurred in main: {e}')

if __name__ == '__main__':
    main()
```

2. Write a program to draw swarm plot of “total bill” against day for a dataset given in url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv


```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import logging

''' 
    @Author: v sanjay kumar
    @Date: 2024-08-13 04:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-13 04:00:30.
    @Title :program to draw swarm plot of “total bill” against day for a dataset given in url.
'''

logging.basicConfig(filename='app2.log', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def create_swarm_plot(url):
    '''
    Function takes the parameter from the url and plots the swarm plot

    Parameters:
    url (str): The URL of the dataset

    Returns:
    None
    '''
    try:
        df = pd.read_csv(url)
        logging.info('Dataset loaded successfully.')

        plt.figure(figsize=(10, 6))
        sns.swarmplot(x='day', y='total_bill', data=df)
        logging.info('Swarm plot created successfully.')

        plt.xlabel('Day')
        plt.ylabel('Total Bill')
        plt.title('Swarm Plot of Total Bill by Day')
        logging.info('Labels and title added to the plot.')

        plt.show()
        logging.info('Plot displayed successfully.')

    except Exception as e:
        logging.error(f'Error occurred: {e}')

def main():
    try:
        url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
        create_swarm_plot(url)
    
    except Exception as e:
        logging.error(f'Error occurred in main: {e}')

if __name__ == '__main__':
    main()
```
