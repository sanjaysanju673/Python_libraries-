### 3. Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.

```python
def plot_line(x, y):
    '''
    Function takes the x, y points and plots a graph.

    Parameters:
    x (List[int]): List of x points.
    y (List[int]): List of y points.

    Returns:
    None
    '''
    plt.plot(x, y, color='r', linewidth=5)
    plt.xlabel('X Axis values')
    plt.ylabel('Y Axis values')
    plt.title('Line values')
    plt.grid(axis='y')
    plt.show()

def main():
    x = []
    y = []
    with open('model.txt', 'r') as file:
        next(file)  
        for line in file:
            values = line.split()
            x.append(int(values[0]))
            y.append(int(values[1]))
    
    plot_line(x, y)

if __name__ == '__main__':
    main()
```

### 9. Write a Python program to display the current axis limits values and set new axis values.

```python

def change_axis(x,y):
    '''
        function takes the x,y points and plot the one graph

        parameters(List,List)-takes x points list and y points list

        retun -None
    '''
    plt.plot(x, y)
    current_limits = plt.axis()
    print("Current axis limits:", current_limits)
    plt.axis([0, 6, 0, 50])
    new_limits = plt.axis()
    print("New axis limits:", new_limits)
    plt.xlabel('x - axis')
    plt.ylabel('y - axis')
    plt.title('Setting new axis limits')
    plt.show()


def main():
    x = [1, 2, 3, 4, 5]
    y = [10, 20, 25, 30, 40]
    change_axis(x,y)


if __name__ == '__main__':
    main()



```

### 14. Write a Python program to display the grid and draw line charts of the closing value ofAlphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines withrendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turnoff ticks.

```python

def main():
    dates = ['03-10-16', '04-10-16', '05-10-16', '06-10-16', '07-10-16']
    closing_values = [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]

    dates = [datetime.datetime.strptime(date, '%d-%m-%y') for date in dates]

    plt.plot(dates, closing_values, marker='o', linestyle='-', color='b')


    plt.grid(True, which='both', linestyle='-', linewidth=0.5, color='blue')


    plt.minorticks_on()
    plt.grid(which='major', linestyle='-', linewidth='0.5', color='blue')
    plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
    plt.tick_params(which='both', top=False, bottom=False, left=False, right=False)

    plt.xlabel('Date')
    plt.ylabel('Closing Value')
    plt.title('Closing Value of Alphabet Inc. (Oct 3, 2016 - Oct 7, 2016)')

    plt.show()


if __name__ =='__main__':
    main()
```

### 15. Write a Python program to create multiple plots.

```python
def main():
    x = np.linspace(0, 2 * np.pi, 400)
    y = np.sin(x ** 2)

    fig, axs = plt.subplots(2, 2)


    axs[0, 0].plot(x, y)
    axs[0, 0].set_title('Plot 1: sin(x^2)')

    axs[0, 1].plot(x, np.cos(x ** 2), 'tab:orange')
    axs[0, 1].set_title('Plot 2: cos(x^2)')

    axs[1, 0].plot(x, -y, 'tab:green')
    axs[1, 0].set_title('Plot 3: -sin(x^2)')

    axs[1, 1].plot(x, -np.cos(x ** 2), 'tab:red')
    axs[1, 1].set_title('Plot 4: -cos(x^2)')

    for ax in axs.flat:
        ax.set(xlabel='x-label', ylabel='y-label')

    for ax in axs.flat:
        ax.label_outer()

    plt.show()

if __name__ == '__main__':
    main()

    
    
   
```
