1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python

'''
    @Author: v sanjay kumar
    @Date: 2024-08-12 10:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-12 10:00:30.
    @Title :program to  draw a line with suitable label in the x axis, y axis and a title.

'''


def plot_graph(x,y):
    '''
        function takes the x, y points and plot the one graph

        parameters(List,List)-takes x points list and y points list

        retun -None
    '''
    plt.plot(x,y,linestyle='dotted',color='green')
    plt.title('Total Sales of the years')
    plt.xlabel('years')
    plt.ylabel('sales')
    plt.show()
    


def main():
    y=[10000,22345,423456,524689,712342,83453,912432]
    x=[5,3,7,5,3,2,2]
    plot_graph(x,y)


if __name__ == '__main__':
    main()
```

2. Write a Python program to draw a line using given axis values with suitable label in the x axis
, y axis and a title

```python
'''
    @Author: v sanjay kumar
    @Date: 2024-08-12 10:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-12 10:00:30.
    @Title :program to  draw a line with suitable label in the x axis, y axis and a title.

'''


def main():
    x=[1,2,3,4,5,6,7]
    y=[1,2,3,4,5,6,7]
    plt.plot(x, y, color='y')
    plt.xlabel('X axis values')
    plt.ylabel('Y axis values')
    plt.title('Line plot with given axies Values')
    plt.show()

if __name__ == '__main__':
    main()
```
