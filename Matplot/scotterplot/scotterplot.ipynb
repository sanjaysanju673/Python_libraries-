1. Write a Python program to draw a scatter graph taking a random distribution in X and Y and
plotted against each other.

```python
import matplotlib.pyplot as plt
import numpy as np
'''
    @Author: v sanjay kumar
    @Date: 2024-08-12 06:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-12 06:00:30.
    @Title :program to  draw a scatter graph taking a random distribution in X and Y and
           plotted against each other. 
'''


def scatter_data(x,y):
    '''
    Function to draw a scatter graph taking a random distribution in X and Y plotted against each other.
    
    parameters-list,list=x,y
    
    return -None
    '''
    plt.scatter(x, y, c ="pink", 
                linewidths = 2, 
                marker ="s", 
                edgecolor ="green",s = 50)

    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()

def main():
    x = np.random.randint(10,50,10)
    y =np.random.randint(10,50,10)

    scatter_data(x,y)

if __name__ =='__main__':
    main()
```

3. Write a Python program to draw a scatter plot using random distributions to generate balls of
different sizes.

```python
import matplotlib.pyplot as plt

import math
import random
'''
    @Author: v sanjay kumar
    @Date: 2024-08-12 06:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-12 06:00:30.
    @Title :program to draw a scatter plot with empty circles taking a random distribution
        in X and Y and plotted against each other. 
'''
def color_plot(x,y,colors,areas):
    ''' function takes x,y colors and take and plot the scotter.
       
    parameters(list,list,list,list)-x,y,colors,areas.

    return-None
    '''
    plt.figure()
    plt.scatter(x, y, s=areas, c=colors, alpha=0.85)
    plt.axis([0.0, 1.0, 0.0, 1.0])
    plt.title('scatter plot with empty circles with different colors')
    plt.xlabel("X axis",rotation=0)
    plt.ylabel("Y axis",rotation=0)




def main():
    no_of_balls = 25
    x = [random.triangular() for i in range(no_of_balls)]
    y = [random.gauss(0.5, 0.25) for i in range(no_of_balls)]
    colors = [random.randint(1, 4) for i in range(no_of_balls)]
    areas = [math.pi * random.randint(5, 15)**2 for i in range(no_of_balls)]
    color_plot(x,y,colors,areas)
    

if __name__ =='__main__':
    main()
```
