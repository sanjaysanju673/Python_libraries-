1. Write a program to draw a point plot for sex against survived for a dataset given in url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
''' 
    @Author: v sanjay kumar
    @Date: 2024-08-12 06:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-12 06:00:30.
    @Title :program to draw a point plot for sex against survived for a dataset given in url.
'''
def plot_data(url):
    '''
    function takes url from the main function and returns point plot

    paramatar(url)-take the url from main function

    Return-None
    '''
    df=pd.read_csv(url)
    sns.pointplot(x='sex', y='survived', data=df,color=(0.9,0.6,0.9),linewidth=3)
    plt.xlabel('Sex')
    plt.ylabel('Survived')
    plt.title('Survival Rate by Sex')
    plt.show()




def main():
    try:

        url='https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv'
        plot_data(url)
    
    except Exception as e:
        print('Invalid Input',e)


if __name__ == '__main__':
    main()
```
