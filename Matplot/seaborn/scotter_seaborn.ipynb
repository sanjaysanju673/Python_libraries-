1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a
url

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import logging as log

''' 
    @Author: v sanjay kumar
    @Date: 2024-08-12 06:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-12 06:00:30.
    @Title :program to draw a scatter plot of “day” against “total bill” for a dataset given in a url
'''
log.basicConfig(filename='app.log', level=log.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')
def scotter_plot(url):
    '''
    Function takes the git hub url and show scotter plot

    parameters-url -git hub url

    returns-None
    '''
    log.info('cutsor enter into function')
    tips = pd.read_csv(url)

    sns.scatterplot(x='day', y='total_bill', data=tips)

    plt.xlabel('Day')
    plt.ylabel('Total Bill')
    plt.title('Scatter Plot of Day vs Total Bill')
    log.warning("thise is warnig mesage")
    plt.show()
    


def main():
    try:
        url = 'https://raw.githubusercontent.om/mwaskom/seaborn-data/master/tips.csv'
        scotter_plot(url)
        log.info('The cutsor enter in the main function')
    except Exception as e:
        log.error('Thise is error mesage',e)

if __name__ == '__main__':
    main()
       
    
```
