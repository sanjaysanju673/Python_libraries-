1. Write a Python programming to create a pie chart of the popularity of programming
Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt
'''
    @Author: v sanjay kumar
    @Date: 2024-08-12 06:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-12 06:00:30.
    @Title :program to create a pie chart of the popularity of programming.
'''

def Pie_plot(programming_language,popularity):
    '''
    function take the programming_language and popularity list and print pie chart
    parameters list,list-programming_language,popularity
    Return-None
    '''
    plt.figure(figsize=(10,10))
    plt.pie(popularity,labels=programming_language)
    plt.legend()
    plt.show()
    

def main():
    programming_language=[ 'Java','Python', 'PHP', 'JavaScript', 'C#', 'C++']
    Popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
    Pie_plot(programming_language,Popularity)
    
```

2. Write a Python programming to create a pie chart with a title of the popularity of programming
Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt
'''
    @Author: v sanjay kumar
    @Date: 2024-08-12 06:00:30
    @Last Modified by: v sanjay kumar
    @Last Modified time: 2024-08-12 06:00:30.
    @Title :programto create a pie chart with a title of the popularity of programming Languages.
'''

def Pie_plot(programming_language,popularity):
    '''
    function take the programming_language and popularity list and print pie chart

    parameters list,list-programming_language,popularity

    Return-None
    '''
    plt.figure(figsize=(8,8))
    plt.pie(popularity,labels=programming_language)
    plt.title(' The Popularity Of Programming Languages')
    plt.legend()
    plt.show()
    

def main():
    programming_language=[ 'Java','Python', 'PHP', 'JavaScript', 'C#', 'C++']
    Popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
    Pie_plot(programming_language,Popularity)


if __name__ == '__main__':
    main()
        
```
