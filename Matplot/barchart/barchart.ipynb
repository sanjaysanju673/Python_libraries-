1. Write a Python programming to display a bar chart of the popularity of programming
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>


```python
def plot_bar(languages,popularity,):
        '''
        the function takes languages and popularity and plot the bargraph

        parameters list,list-languages,popularity

        returns-None
        '''

        
        x_pos = [i for i in range(len(languages))]
        plt.bar(x_pos, popularity, color='green' ,width=0.4)

        plt.xlabel('Programming Languages')
        plt.ylabel('Popularity')
        plt.title('Popularity of Programming Languages')
        plt.xticks(x_pos, languages)
        plt.show()
    
def main():

        languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
        popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
        plot_bar(languages, popularity)

if __name__ == '__main__':
    main()
```

### 9.Write a Python programming to display a bar chart of the popularity of programmingLanguages. Increase bottom margin.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>

```python
def plot_data():
    ''' Function to display a bar chart of the popularity of programming
        Languages. Select the width of each bar and their positions.

        parameteres-None

        return -None
    '''
    languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
        
    x_pos = [i for i in range(len(languages))]
    widths =0.4

    plt.bar(x_pos, popularity, color='orange', width=widths)

    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity')
    plt.title('Popularity of Programming Languages')

    plt.xticks(x_pos, languages)
    plt.subplots_adjust(bottom=0.4)

    plt.show()

def main():
    plot_data()


if __name__ == '__main__':
    main()
```

### 10. Write a Python program to create bar plot of scores by group and gender. Use multiple Xvalues on the same chart for men and women.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>

```python
def plot_bar():
    '''
    Function to display a bar chart of the popularity of programming
        Languages. Select the width of each bar and their positions

    parameters-None

    return-None
    '''
    means_men = [22, 30, 35, 35, 26]
    means_women = [25, 32, 30, 35, 29]
    groups = ['G1', 'G2', 'G3', 'G4', 'G5']


    n_groups = len(groups)

    bar_width = 0.35

    opacity = 0.8

    index = np.arange(n_groups)

    rects1 = plt.bar(index, means_men, bar_width, alpha=opacity, color='b', label='Men')

    rects2 = plt.bar(index + bar_width, means_women, bar_width, alpha=opacity, color='r', label='Women')

    plt.xlabel('Group')
    plt.ylabel('Scores')
    plt.title('Scores by group and gender')
    plt.xticks(index + bar_width / 2, groups)
    plt.legend()

  
    plt.tight_layout()
    plt.show()


def main():
    plot_bar()


if __name__ == '__main__':
    main()
```

### 12. Write a Python program to create bar plots with error bars on the same figure. Sample Date
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824 <br>
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645<br>

```python

def plot_error(mean_velocity,std_deviation,x_labels):
    '''
    function  create bar plots with error bars on the same figure.
.

    parameters list,list,list-mean_velocity,std_deviation,x_labels.

    Return-None
    '''
    
    

    plt.figure(figsize=(9,5))

    plt.bar(x_labels, mean_velocity, yerr=std_deviation, capsize=4, color='skyblue', ecolor='black')


    plt.xlabel('Samples')
    plt.ylabel('Mean Velocity')
    plt.title('Bar Plot with Error Bars')

    plt.show()

def main():
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]
    x_labels = ['A', 'B', 'C', 'D']
    plot_error(mean_velocity,std_deviation,x_labels)


if __name__ == '__main__':
    main()


```

13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text
label above each bar displaying men means (integer value).


```python
import logging
logging.basicConfig(filename="txt2.txt",level=logging.INFO)

def int_plot(mean_velocity, std_deviation, x_labels):
    '''
    function take the mean_velocity and std_deviation and x_labels and print the 

    parameters list,list,list-mean_velocity,std_deviation,x_labels.

    Return-None
    '''
    logging.info("Creating bar plot with error bars.")
    
    plt.figure(figsize=(10, 6))
    bars = plt.bar(x_labels, mean_velocity, yerr=std_deviation, capsize=5, color='skyblue', ecolor='black')

    for bar, mean in zip(bars, mean_velocity):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.02, f'{int(mean*1000)}', ha='center', va='bottom')
        logging.info(f"Bar label for {bar.get_x()}: {int(mean*1000)}")

    plt.xlabel('Samples')
    plt.ylabel('Mean Velocity')
    plt.title('Bar Plot with Error Bars and Mean Labels')

    plt.show()
    logging.info("Bar plot displayed successfully.")

def main():
    logging.info("Program started.")
    
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]
    x_labels = ['A', 'B', 'C', 'D']
    
    int_plot(mean_velocity, std_deviation, x_labels)
    
    logging.info("Program finished.")

if __name__ == "__main__":
    main()
```
